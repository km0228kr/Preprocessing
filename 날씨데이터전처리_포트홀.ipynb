    - 확인 결과 'Unnaemd: 0'이라는 컬럼이 추가되어있는 구가 있으므로 해당 컬럼이 있는 경우에는 삭제
    - 일교차 컬럼 생성

```python
import pandas as pd
```

```python
def gu_merge(gu_list):
    # 파일 경로 지정
    p_url = 'C:\\JupyterNotebook\\COIN\\PotHole\\1 데이터\\1-1 포트홀데이터\\포트홀데이터전처리.csv'
    w_url = 'C:\\JupyterNotebook\\COIN\\PotHole\\1 데이터\\\\1-2 구별 날씨데이터_시간_완\\{}기상데이터.csv'
    
    
    # 포트홀 데이터
    p_df = pd.read_csv(p_url)

    for i in range(len(gu_list)):
        # 날씨 데이터
        w_df = pd.read_csv(w_url.format(gu_list[i]))
        print(w_url.format(gu_list[i]))


        # 보수 날짜 기준 그룹화
        counter = pd.DataFrame(p_df.groupby(by='보수날짜')['구'].count())
        square = pd.DataFrame(p_df.groupby(by='보수날짜')['파손면적'].sum()) 

        counter.reset_index(inplace=True)
        square.reset_index(inplace=True)

        counter.columns = ['Date','발생건수']
        square.columns = ['Date','파손면적합계']


        # 포트홀 데이터와 날씨 데이터 합치기
        w_df = pd.merge(w_df,counter, how='outer', on='Date')
        w_df = pd.merge(w_df,square, how='outer', on='Date')

        # 결측치 처리
        w_df['발생건수'].fillna(0,inplace=True)
        w_df['파손면적합계'].fillna(0, inplace=True)

        
        # Unnamed: 0 컬럼이 있는 구가 있으므로 해당 컬럼이 있으면 삭제
        if 'Unnamed: 0' in w_df.columns:
            w_df = w_df.drop('Unnamed: 0', axis=1)

        # Index룰 Date로 설정
        w_df.set_index('Date',inplace=True)
        
        # 일교차 컬럼 만들기
        daily_temp = w_df['최고기온'] - w_df['최저기온']
        w_df['일교차'] = daily_temp        

        # 개별 파일로 저장
        w_df.to_csv('C:\\JupyterNotebook\\COIN\\PotHole\\1 데이터\\1-3 포트홀_날씨_merge\\{}_포트홀_날씨.csv'.format(gu_list[i]), encoding='utf-8-sig')
```
